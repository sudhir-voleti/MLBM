```python
import requests
import pandas as pd

# User input for API key and city
api_key = input("Please enter your OpenWeatherMap API key: ")
city_name = input("Please enter the city name (e.g., Hyderabad): ")

# Construct API query URL
base_url = "https://api.openweathermap.org/data/2.5/weather"
params = {
    'q': city_name,
    'appid': api_key,
    'units': 'metric'  # For temperature in Celsius
}

# Send API request and receive response
try:
    response = requests.get(base_url, params=params)
    response.raise_for_status()  # Raise HTTPError for bad responses (4xx or 5xx)
    weather_data = response.json()

    # Print neat output dataframe
    df = pd.DataFrame({
        'Parameter': [
            'City', 'Country', 'Latitude', 'Longitude', 'Weather Condition',
            'Description', 'Temperature (°C)', 'Feels Like (°C)', 'Humidity (%)',
            'Wind Speed (m/s)', 'Cloudiness (%)', 'Sunrise (UTC)', 'Sunset (UTC)'
        ],
        'Value': [
            weather_data['name'], weather_data['sys']['country'],
            weather_data['coord']['lat'], weather_data['coord']['lon'],
            weather_data['weather'][0]['main'], weather_data['weather'][0]['description'],
            weather_data['main']['temp'], weather_data['main']['feels_like'],
            weather_data['main']['humidity'], weather_data['wind']['speed'],
            weather_data['clouds']['all'], weather_data['sys']['sunrise'],
            weather_data['sys']['sunset']
        ]
    })

    print(df.to_string(index=False))

except requests.exceptions.HTTPError as http_err:
    print(f"HTTP error occurred: {http_err}")
except Exception as err:
    print(f"An error occurred: {err}")
```

## Third prompt featuring a variation (P3)
Nice, it works. Now, say I have a list of major cities in the world {Tokyo, Beijing, Delhi, Moscow, Paris, New York}.
I'd like to collect current weather data on each and put into a dataframe. Modify code to enable this pls.


```python
import requests
import pandas as pd

# User input for API key
api_key = input("Please enter your OpenWeatherMap API key: ")

# List of major cities
cities = ["Tokyo", "Beijing", "Delhi", "Moscow", "Paris", "New York"]

# Base URL for OpenWeatherMap API
base_url = "https://api.openweathermap.org/data/2.5/weather"

# List to store weather data for all cities
all_cities_data = []

# Loop through each city and fetch weather data
for city_name in cities:
    params = {
        'q': city_name,
        'appid': api_key,
        'units': 'metric'  # For temperature in Celsius
    }

    try:
        response = requests.get(base_url, params=params)
        response.raise_for_status()
        weather_data = response.json()

        # Extract relevant weather parameters for each city
        city_data = {
            'City': weather_data['name'],
            'Country': weather_data['sys']['country'],
            'Latitude': weather_data['coord']['lat'],
            'Longitude': weather_data['coord']['lon'],
            'Weather Condition': weather_data['weather'][0]['main'],
            'Description': weather_data['weather'][0]['description'],
            'Temperature (°C)': weather_data['main']['temp'],
            'Feels Like (°C)': weather_data['main']['feels_like'],
            'Humidity (%)': weather_data['main']['humidity'],
            'Wind Speed (m/s)': weather_data['wind']['speed'],
            'Cloudiness (%)': weather_data['clouds']['all'],
            'Sunrise (UTC)': weather_data['sys']['sunrise'],
            'Sunset (UTC)': weather_data['sys']['sunset']
        }
        all_cities_data.append(city_data)

    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP error occurred for {city_name}: {http_err}")
    except Exception as err:
        print(f"An error occurred for {city_name}: {err}")

# Create DataFrame from the collected data
if all_cities_data: # Check if there is any data to create dataframe
    df_all_cities = pd.DataFrame(all_cities_data)
    print(df_all_cities.to_string(index=False))
else:
    print("No weather data could be retrieved for any of the cities.")
```

```python
import requests
import json

# Prompt for API key and keywords
api_key = input("Enter your NewsAPI API Key: ")
keywords = input("Enter keywords for your news search: ")

# Base URL for the NewsAPI everything endpoint
base_url = "https://newsapi.org/v2/everything"

# Parameters for the request
params = {
    'q': keywords,
    'apiKey': api_key
}

try:
    # Send GET request
    response = requests.get(base_url, params=params)
    response.raise_for_status() # Raise an exception for HTTP errors

    # Parse JSON response
    data = response.json()

    # Check for successful status
    if data['status'] == 'ok':
        print("Total articles found:", data['totalResults'])
        articles = data['articles']

        if articles:
            print("\nArticles:\n")
            for article in articles:
                print(f"Title: {article['title']}")
                print(f"URL: {article['url']}")
                print("-" * 40)
        else:
            print("No articles found matching your keywords.")
    else:
        print("Error in API request:")
        print(data) # Print the error response for debugging

except requests.exceptions.HTTPError as http_err:
    print(f"HTTP error occurred: {http_err}")
except Exception as err:
    print(f"An error occurred: {err}")
```

```python
import requests
import pandas as pd
import json
from IPython.display import HTML
from google.colab import files

def get_news_dataframe(api_key, keywords):
    """
    Fetches news articles from NewsAPI based on keywords and returns a Pandas DataFrame.

    Args:
        api_key: Your NewsAPI API key.
        keywords: Keywords to search for in news articles.

    Returns:
        pandas.DataFrame: DataFrame containing news articles, or None if an error occurs.
    """
    base_url = "https://newsapi.org/v2/everything"
    params = {
        "q": keywords,
        "apiKey": api_key,
        "pageSize": 100  # Increased page size to get more articles per request
    }
    articles_list = []
    try:
        response = requests.get(base_url, params=params)
        response.raise_for_status()
        data = response.json()

        if data['status'] == 'ok':
            articles = data['articles']
            for article in articles:
                articles_list.append({
                    'Title': article['title'],
                    'Description': article['description'],
                    'Source Name': article['source']['name'],
                    'Author': article['author'],
                    'Published At': article['publishedAt'],
                    'URL': article['url']
                })
        else:
            print("Error in API request:")
            print(data)
            return None

    except requests.exceptions.RequestException as e:
        print(f"Error fetching news: {e}")
        return None

    df = pd.DataFrame(articles_list)
    return df

# 1. Prompt user for API Key and Keywords
api_key = input("Enter your NewsAPI API Key: ")
keywords = input("Enter keywords for your news search (e.g., Trump Zelensky): ")

# 2. Fetch News Data and Create DataFrame
news_df = get_news_dataframe(api_key, keywords)

if news_df is not None:
    # 3. Display Top 12 Rows as HTML Table
    top_12_df = news_df.head(12) # Get the first 12 rows
    html_table = top_12_df.to_html(escape=False, index=False)
    display(HTML(html_table))

    # 4. Option to Download as CSV
    csv_data = news_df.to_csv(index=False)
    download_button = files.download(csv_data.encode('utf-8'), filename='news_articles.csv') # Corrected download call
    print("\nDownload CSV file (all rows) by clicking the button below:")
    display(download_button)

else:
    print("Could not retrieve news data.")
```

```python
import requests
import pandas as pd
from IPython.display import display, HTML

# Function to extract video ID from URL
def extract_video_id(url):
    """
    Extracts the video ID from a YouTube URL.
    Example: https://www.youtube.com/watch?v=4GLSzuYXh6w -> 4GLSzuYXh6w
    """
    from urllib.parse import urlparse, parse_qs
    parsed_url = urlparse(url)
    if parsed_url.hostname == 'youtu.be':  # Shortened URL format
        return parsed_url.path[1:]
    if 'v' in parse_qs(parsed_url.query):  # Standard URL format
        return parse_qs(parsed_url.query)['v'][0]
    return None

# Get user input for the API key
api_key = input("Enter your YouTube Data API key: ")

# Get user input for the YouTube video URL
video_url = input("Enter the YouTube video URL: ")
video_id = extract_video_id(video_url)

if not api_key:
    print("Error: API key is required.")
elif not video_id:
    print("Invalid YouTube URL. Please provide a valid URL.")
else:
    # Construct the API request URL
    url = f"https://www.googleapis.com/youtube/v3/videos?part=snippet,contentDetails,statistics&id={video_id}&key={api_key}"

    # Send GET request
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        data = response.json()

        # Extract relevant information
        if data['items']:
            video = data['items'][0]
            snippet = video['snippet']
            statistics = video['statistics']

            # Create a dictionary for the video details
            video_details = {
                "Title": snippet.get('title', 'N/A'),
                "Channel": snippet.get('channelTitle', 'N/A'),
                "Published Date": snippet.get('publishedAt', 'N/A'),
                "Views": statistics.get('viewCount', 'N/A'),
                "Likes": statistics.get('likeCount', 'N/A'),
                "Comments": statistics.get('commentCount', 'N/A'),
                "Duration": video['contentDetails'].get('duration', 'N/A')
            }

            # Convert to DataFrame
            df = pd.DataFrame([video_details])

            # Display as an HTML table
            print("Video Details:")
            display(HTML(df.to_html(index=False)))
        else:
            print("No data found for the provided video ID.")
    else:
        print(f"Error: Unable to fetch data (Status Code: {response.status_code})")
        print(response.text)
```

```python
import requests
import pandas as pd
from IPython.display import display, HTML

# Function to extract video ID from URL
def extract_video_id(url):
    """
    Extracts the video ID from a YouTube URL.
    Example: https://www.youtube.com/watch?v=4GLSzuYXh6w -> 4GLSzuYXh6w
    """
    from urllib.parse import urlparse, parse_qs
    parsed_url = urlparse(url)
    if parsed_url.hostname == 'youtu.be':  # Shortened URL format
        return parsed_url.path[1:]
    if 'v' in parse_qs(parsed_url.query):  # Standard URL format
        return parse_qs(parsed_url.query)['v'][0]
    return None

# Get user input for the API key
api_key = input("Enter your YouTube Data API key: ")

# Get user input for the YouTube video URL
video_url = input("Enter the YouTube video URL: ")
video_id = extract_video_id(video_url)

if not api_key:
    print("Error: API key is required.")
elif not video_id:
    print("Invalid YouTube URL. Please provide a valid URL.")
else:
    # Step 1: Fetch video details
    video_details_url = f"https://www.googleapis.com/youtube/v3/videos?part=snippet,contentDetails,statistics&id={video_id}&key={api_key}"
    response = requests.get(video_details_url)

    if response.status_code != 200:
        print(f"Error fetching video details (Status Code: {response.status_code})")
        print(response.text)
    else:
        data = response.json()

        if not data['items']:
            print("No data found for the provided video ID.")
        else:
            video = data['items'][0]
            snippet = video['snippet']
            statistics = video['statistics']

            # Create a dictionary for the video details
            video_details = {
                "Title": snippet.get('title', 'N/A'),
                "Channel": snippet.get('channelTitle', 'N/A'),
                "Published Date": snippet.get('publishedAt', 'N/A'),
                "Views": statistics.get('viewCount', 'N/A'),
                "Likes": statistics.get('likeCount', 'N/A'),
                "Comments": statistics.get('commentCount', 'N/A'),
                "Duration": video['contentDetails'].get('duration', 'N/A')
            }

            # Display video details
            df_video = pd.DataFrame([video_details])
            print("Video Details:")
            display(HTML(df_video.to_html(index=False)))

            # Step 2: Fetch comments
            comments_url = f"https://www.googleapis.com/youtube/v3/commentThreads?part=snippet&videoId={video_id}&key={api_key}&maxResults=10"
            comments_response = requests.get(comments_url)

            if comments_response.status_code != 200:
                print(f"Error fetching comments (Status Code: {comments_response.status_code})")
                print(comments_response.text)
            else:
                comments_data = comments_response.json()

                if not comments_data['items']:
                    print("No comments found for this video.")
                else:
                    # Extract comment details
                    comments_list = []
                    for item in comments_data['items']:
                        comment_snippet = item['snippet']['topLevelComment']['snippet']
                        comments_list.append({
                            "Author": comment_snippet.get('authorDisplayName', 'N/A'),
                            "Comment": comment_snippet.get('textDisplay', 'N/A'),
                            "Published At": comment_snippet.get('publishedAt', 'N/A'),
                            "Like Count": comment_snippet.get('likeCount', 'N/A')
                        })

                    # Convert to DataFrame
                    df_comments = pd.DataFrame(comments_list)

                    # Display comments
                    print("\nTop Comments:")
                    display(HTML(df_comments.to_html(index=False)))
```

```python
import requests
import pandas as pd
from datetime import datetime, timedelta
from IPython.display import display, HTML

# Function to extract article title from Wikipedia URL
def extract_article_title(url):
    """
    Extracts the article title from a Wikipedia URL.
    Example: https://en.wikipedia.org/wiki/Indian_School_of_Business -> Indian_School_of_Business
    """
    from urllib.parse import urlparse
    parsed_url = urlparse(url)
    if parsed_url.hostname.endswith('wikipedia.org') and parsed_url.path.startswith('/wiki/'):
        return parsed_url.path.split('/wiki/')[-1]
    return None

# Get user input for the Wikipedia URL
wiki_url = input("Enter the Wikipedia URL: ")
article_title = extract_article_title(wiki_url)

if not article_title:
    print("Invalid Wikipedia URL. Please provide a valid URL.")
else:
    # Define the Wikimedia Pageviews API endpoint
    api_url = "https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/{project}/{access}/{agent}/{article}/daily/{start_date}/{end_date}"

    # Parameters
    project = "en.wikipedia"  # English Wikipedia
    access = "all-access"     # All access methods (desktop, mobile, etc.)
    agent = "user"            # Exclude bots
    start_date = (datetime.now() - timedelta(days=30)).strftime('%Y%m%d')  # Start date: 30 days ago
    end_date = datetime.now().strftime('%Y%m%d')                           # End date: today

    # Construct the API request URL
    url = api_url.format(
        project=project,
        access=access,
        agent=agent,
        article=article_title,
        start_date=start_date,
        end_date=end_date
    )

    # Send GET request
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
    response = requests.get(url, headers=headers)

    # Check if the request was successful
    if response.status_code != 200:
        print(f"Error fetching page views (Status Code: {response.status_code})")
        print(response.text)
    else:
        data = response.json()

        if 'items' not in data:
            print("No page view data found for this article.")
        else:
            # Extract daily page views
            page_views = []
            for item in data['items']:
                date = item['timestamp'][:8]  # Extract YYYYMMDD
                views = item['views']
                page_views.append({
                    "Date": datetime.strptime(date, '%Y%m%d').strftime('%Y-%m-%d'),
                    "Views": views
                })

            # Convert to DataFrame
            df = pd.DataFrame(page_views)

            # Display as an HTML table
            print(f"Daily Page Views for '{article_title.replace('_', ' ')}' (Past Month):")
            display(HTML(df.to_html(index=False)))
```

Write python code for colab that [1] asks for and accepts user inputs viz. API key,
[2] then creates a chatbot that asks for and accepts user input in natural language on target consumer profile or preferences (e.g., "young professionals interested in fitness"),
[3] and finally, shows recommended items for target consumer as a dataframe with 2 columns - 'items recommended' and 'reasoning'.

Note how we structured the input prompt further this time into numbered subtasks. Note also how it is a good idea to supply the URL of the API documentation page.

Had to do an iteration or two to debug errors and the final code I got is below:

```python
# Import necessary libraries
import requests
import pandas as pd

# [1] Accept API Key from the user
API_KEY = input("Please enter your Mistral API key: ").strip()
API_URL = "https://api.mistral.ai/v1/chat/completions"  # Correct endpoint from Mistral API docs

# Function to call the Mistral API
def call_mistral_api(prompt, max_tokens=200):
    headers = {
        "Authorization": f"Bearer {API_KEY}",
        "Content-Type": "application/json"
    }
    data = {
        "model": "mistral-small",  # Specify the model (adjust based on available models)
        "messages": [
            {"role": "user", "content": prompt}
        ],
        "max_tokens": max_tokens
    }
    response = requests.post(API_URL, headers=headers, json=data)
    if response.status_code == 200:
        return response.json().get("choices", [{}])[0].get("message", {}).get("content", "No response generated.")
    else:
        print(f"Error: {response.status_code}, {response.text}")
        return None

# [2] Create a chatbot to accept user input about target consumer profile/preferences
def get_user_input():
    print("\nWelcome to the Personalized Content Generator!")
    print("Please describe the target consumer profile or preferences in natural language.")
    user_input = input("Example: 'Young professionals interested in fitness': ")
    return user_input.strip()

# Generate recommendations based on user input
def generate_recommendations(user_input):
    # Construct the prompt for the LLM
    prompt = (
        f"Based on the following target consumer profile or preferences: '{user_input}', "
        f"recommend 5 items or services and provide reasoning for each recommendation. "
        f"Format the output as a list of items with their reasoning, separated by colons (:)."
    )

    # Call the Mistral API
    response = call_mistral_api(prompt, max_tokens=200)
    if not response:
        return None

    # Parse the response into a DataFrame
    try:
        # Split the response into items and reasoning
        lines = response.split("\n")
        items = []
        reasoning = []
        for line in lines:
            if ":" in line:
                item, reason = line.split(":", 1)
                items.append(item.strip())
                reasoning.append(reason.strip())

        # Create a DataFrame
        df = pd.DataFrame({
            "items recommended": items,
            "reasoning": reasoning
        })
        return df
    except Exception as e:
        print(f"Error parsing response: {e}")
        return None

# Main execution flow
if __name__ == "__main__":
    # Step 1: Get API key
    if not API_KEY:
        print("API key is required to proceed.")
    else:
        # Step 2: Get user input
        user_input = get_user_input()

        # Step 3: Generate recommendations
        recommendations_df = generate_recommendations(user_input)

        # Display results
        if recommendations_df is not None:
            print("\nHere are the personalized recommendations:")
            display(recommendations_df)
        else:
            print("Failed to generate recommendations. Please try again.")
```
